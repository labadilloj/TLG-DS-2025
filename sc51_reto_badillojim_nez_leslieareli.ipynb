###**Reto | Autotransporte Federal de Turismo**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D  # para la gráfica 3D
import statsmodels.api as sm
import numpy as np
```

```python
df_1 = pd.read_csv('reto_1.csv', encoding='latin-1')
df_3 = pd.read_csv('reto_3.csv', encoding='latin-1')
```

```python
print(df_1.head())
print(df_3.head())
```

####Gráfica de caja de bigotes.

```python
plt.figure()
ax = sns.boxplot(data=df_1, x="Autobuses", color="yellow")
plt.title("Caja de bigotes de Autobuses por mes")
mediana = df_1["Autobuses"].median()
ax.annotate(
        f"Mediana: {mediana:.1f}",
        xy=(mediana, 0),
        xytext=(mediana + 20, 0.5),
        arrowprops=dict(facecolor='black', shrink=0.05),
        horizontalalignment='right',
        verticalalignment='bottom')

plt.show()
```

####Gráfica de tres dimensiones.

```python
fig = plt.figure(figsize=(6, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df_1['Kilómetros'], df_1['Autobuses'], df_1['Pasajeros'], c='Red', marker='o')
ax.set_xlabel('Kilómetros')
ax.set_ylabel('Autobuses')
ax.set_zlabel('Pasajeros')
plt.title('Gráfica tridimensional')
plt.show()
```

####Gráfica residual.

```python
plt.figure()
sns.residplot(data=df_1, x="Pasajeros", y="Autobuses")
plt.title("Gráfica residual")
plt.xlabel("Total de pasajeros")
plt.ylabel("Residuales")
```

####Indicarás la existencia de una relación de ajuste.

```python
plt.figure()
sns.regplot(data=df_1, x="Pasajeros", y="Autobuses", ci=95,
            scatter_kws={'color': 'darkblue'})
plt.title("Relación de ajuste")
plt.xlabel("Pasajeros")
plt.ylabel("Autobuses")

# Cálculo del coeficiente de correlación
corr = df_1["Pasajeros"].corr(df_1["Autobuses"])
print(f"Coeficiente de correlación: {corr:.3f}")

if abs(corr) >= 0.7:
    print("Existe una relación de ajuste fuerte.")
elif abs(corr) >= 0.3:
    print("Existe una relación de ajuste moderada.")
else:
    print("La relación de ajuste es débil o casi nula.")
```

####Realizarás una gráfica de kdeplot.

```python
plt.figure()
sns.kdeplot(data=df_1, x="Pasajeros", fill=True, color='purple', linewidth=4)
plt.title("Kdeplot del número de pasajeros")
plt.xlabel("Pasajeros")
plt.ylabel("Densidad")
```

####Construirás un histograma.

```python
plt.figure()
sns.barplot(x='Clase', y='Autobus', data=df_3,
            palette='viridis', hue='Clase',
            legend=False)
plt.title('Autobuses por clase')
plt.xlabel('Clase')
plt.ylabel('Autobuses')
plt.show()
```

####Crearás una gráfica de pastel.

```python
df_3_2 = df_3.head(5)  #Primeras 5 observaciones

plt.figure()
plt.pie(df_3_2['Camioneta'],
    labels=df_3_2['Entidad'],
    autopct="%1.1f%%"
)
plt.title("Porcentaje de camionetas por entidad")
plt.show()
```

####Generarás una gráfica de dispersión.

```python
sns.set_style('whitegrid')
sns.scatterplot(data = df_1, x = "Kilómetros", y = "Autobuses", color='green')
plt.title('Gráfico de dispersión')
plt.xlabel('Kilómetros')
plt.ylabel('Autobuses')
plt.show()
```
