3. Crea cuatro DataFrames en Pandas, cada uno para cargar el contenido de los diferentes archivos.

```python
df_test = pd.read_csv("Test.csv")
#print(df_test.head())

df_test_2 = pd.read_csv("Test2.csv")
#print(df_test_2.head())

df_train = pd.read_csv("Train.csv")
#print(df_train.head())

df_train_2 = pd.read_csv("Train2.csv")
#print(df_train_2.head())
```

4. Conjunta de manera correcta la información, de tal forma que queden todos los registros en un solo DataFrame que contenga todas las columnas necesarias.

```python
df_test = (
    df_test
    .merge(df_test_2, on='Employee_ID')
    )
df_train = (
    df_train
    .merge(df_train_2, on='Employee_ID')
    )
#print(df_test.head())
#print(df_train.head())

key = 'Employee_ID'

# Limpia encabezados
for df in (df_train, df_test):
    df.columns = (df.columns.astype(str)
                  .str.replace(r'\s+', ' ', regex=True)     # colapsa espacios
                  .str.strip())                             # quita espacios al inicio/fin
data_total = pd.concat([df_test, df_train])
print (data_total.head())
```

5. Para todos los registros que no tienen un valor para un sistema operativo, ya sea de su computadora o de su dispositivo móvil, asígnales el valor “no registrado“.

```python
n_before = data_total.isna().sum().sum()

data_total = data_total.replace(r'^\s*$', pd.NA).fillna({
    'Computer_OS': 'No registrado',
    'Mobile_OS': 'No registrado'
})

n_after = data_total.isna().sum().sum()
rellenados_total = n_before - n_after
print('Celdas rellenadas:', rellenados_total)
```



*   El porcentaje de personas que usan Windows, Mac, Linux o "No registrado" como su sistema operativo en su computadora. Se esperan 4 renglones en esta respuesta.

```python
cats = ['Windows', 'MacOS', 'No registrado']
sistema_compu = (data_total['Computer_OS']
       .value_counts(normalize=True)
       .reindex(cats, fill_value=0)
       .mul(100).round(2))
print("\n".join(f"{k}: {v}" for k, v in sistema_compu.items()))

#sistema_computadora = (data_total['Computer_OS'].value_counts(normalize=True) * 100).round(2)
#print(sistema_computadora)
```



*  El porcentaje de personas que usan iOS, Android o "No registrado" como su sistema operativo en su dispositivo móvil. Se esperan 3 renglones en esta respuesta.


```python
cats = ['iOS', 'Android', 'No registrado']
sistema_movil = (data_total['Mobile_OS']
       .value_counts(normalize=True)
       .reindex(cats, fill_value=0)
       .mul(100).round(2))
print("\n".join(f"{k}: {v}" for k, v in sistema_movil.items()))
```

*   La edad promedio de la gente que utiliza cada sistema operativo de escritorio.

```python
edad_prom = (data_total
             .dropna(subset=['Computer_OS','Age'])
             .groupby('Computer_OS')['Age']
             .mean()
             .round(2))

print(edad_prom)
```


*   El nivel educativo promedio de la gente que utiliza cada sistema operativo de escritorio.

```python
educacion_sistema = (data_total
             .dropna(subset=['Computer_OS','Education_Level'])
             .groupby('Computer_OS')['Education_Level']
             .mean()
             .round(2))

print(educacion_sistema)
```



*   El número total de tickets generados por la gente, con cada sistema operativo de escritorio.

```python
tickets_sistema = (data_total
             .dropna(subset=['Computer_OS','Computer_tickets'])
             .groupby('Computer_OS')['Computer_tickets']
             .sum()
             .round(2))

print(tickets_sistema)
```

7. Descarga de xlsx

```python
with pd.ExcelWriter("DS_C2_SC3_Leslie_Badillo.xlsx", engine="openpyxl") as writer:
    data_total.to_excel(writer, sheet_name="Data", index=False)
    sistema_compu.to_excel(writer,    sheet_name="Sistema_Compu",   index=False)
    sistema_movil.to_excel(writer,        sheet_name="Sistema_Movil",       index=False)
    edad_prom.to_excel(writer,         sheet_name="Edad_Promedio",        index=False)
    educacion_sistema.to_excel(writer,         sheet_name="Educacion_Promedio",        index=False)
    tickets_sistema.to_excel(writer,         sheet_name="Tickets_Sistema",        index=False)

print("Archivo 'DS_C2_SC3_Leslie_Badillo.xlsx' creado con 6 hojas")
```
